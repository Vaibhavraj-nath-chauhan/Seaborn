```python
import seaborn as sns
import random as rd
import pandas as pd
import matplotlib.pyplot as plt

tips_df = sns.load_dataset("tips")

tips_df
```

```python
sns.lineplot(x="size",y="tip",data=tips_df)
plt.show()
```

```python
sns.lineplot(x="size",y="tip",data=tips_df,hue="sex")
plt.show()
```

```python
sns.lineplot(x="size",y="tip",data=tips_df,style="sex")
plt.show()
```

```python
sns.lineplot(x="size",y="tip",data=tips_df,hue="sex",style="sex",palette="twilight_shifted_r")
plt.show()
```
