```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
#normal scatter plot
sns.scatterplot(x= df.age,y=df.fare)
plt.show()
```

```python
#adding hue to plot  --> hue
sns.scatterplot(x= df.age,y=df.fare,hue=df.sex)
plt.show()
```

```python
#adding style to diffrinet catigoary --> style
sns.scatterplot(x= df.age,y=df.fare,hue=df.sex,style=df.who)
plt.show()
```

```python
#increasing size of of symboles --> size, sizes
plt.figure(figsize=(15,10))
sns.scatterplot(x= df.age,y=df.fare,hue=df.sex,style=df.who,size=df.who,sizes=(100,200))
plt.show()
```

```python
#increasing size of of symboles --> size, sizes
plt.figure(figsize=(15,10))
sns.scatterplot(x= df.age,y=df.fare,hue=df.sex,style=df.who,size=df.who,sizes=(100,200),palette="hot")
plt.show()
```

```python
#checking other informations 
plt.figure(figsize=(15,10))
sns.scatterplot(x= df.who,y=df.fare,hue=df.alive,style=df.who,size=df.who,sizes=(100,200))
plt.show()
```
