```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_data = sns.load_dataset("tips")
```

```python
tips_data.head()
```

```python
#simple bar plot
sns.barplot(x=tips_data.size)
plt.show()
```

```python
#creating x,y bar plot
sns.barplot(x=tips_data.day,y=tips_data.total_bill)
plt.show()
```

```python
#diffitrating our data   --> hue
sns.barplot(x="day",y="total_bill",data=tips_data,hue=tips_data.sex)
plt.show()
```

```python
#changing order of x axis --> order
order = ["Thur","Sun","Sat","Fri"]
sns.barplot(x="day",y = "total_bill",data=tips_data,hue="sex",order=order)
plt.show()
```

```python
#changing order of hue  -->  hue_order
hue_order = ["Female","Male"]
sns.barplot(x="day",y = "total_bill",data=tips_data,hue="sex",hue_order=hue_order)
plt.show()
```

```python
#our estimator  --> by default value of y is mean we can change that by estimator papramter  --> estimator
import numpy as np

#ploting max
sns.barplot(x="day",y = "total_bill",data=tips_data,hue="sex",estimator=np.max)
plt.show()

#plotiong highest
sns.barplot(x="day",y = "total_bill",data=tips_data,hue="sex",estimator=np.sum)
plt.show()

#ploting min
sns.barplot(x="day",y = "total_bill",data=tips_data,hue="sex",estimator=np.min)
plt.show()
```

```python
#creating horizantal graphs (x sholud in numerical)  --> orient
sns.barplot(x="total_bill",y = "day",data=tips_data,hue="sex",estimator=np.max,orient="h")
plt.show()
```

```python
# changing color   --> color
plt.figure(figsize=(15,10))
sns.barplot(x="day",y = "total_bill",data=tips_data,hue="sex",color = "k")
plt.show()

#advance
# another method --> palette
sns.barplot(x="day",y = "total_bill",data=tips_data,hue="sex",palette="gist_heat")
plt.show()
```
