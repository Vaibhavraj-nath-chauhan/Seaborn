```python
import seaborn as sns
import random as rd
import pandas as pd
import matplotlib.pyplot as plt

tips_df = sns.load_dataset("tips")

tips_df
```

```python
sns.distplot(tips_df.tip)
plt.show()
```

```python
sns.distplot(tips_df.tip,kde=False)
plt.show()
```

```python
sns.distplot(tips_df.tip,hist=False)
plt.show()
```

```python
#color changing 
sns.distplot(tips_df.tip,color="r")
plt.show()
```

```python
# advance color changing 
plt.figure(figsize=(20,10))
sns.distplot(tips_df.tip,
             hist_kws={"color":"k",
                        "edgecolor":"r",
                        "linewidth":3,
                        "linestyle":"--","alpha":0.6},
             kde_kws = {"color":"k",
                        "linewidth":3,
                        "linestyle":"--","alpha":0.3}
            )
```
